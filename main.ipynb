```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')   # downloads the sentence tokenizer model
```

```python
corpus = """Technology is evolving at an unprecedented pace, reshaping industries and daily life. 
            From Generative AI to robotics, innovations are creating new opportunities for learners and professionals alike. 
            Staying updated with these trends not only boosts career prospects but also helps in contributing to impactful projects like Google Summer of Code. 
            With consistent effort, anyone can build skills that align with the future of work."""

#Tokenization
from nltk.tokenize import sent_tokenize,word_tokenize

sents = sent_tokenize(corpus)
for sent in sents:
    print(sent)

words = word_tokenize(corpus)
print(words)
```

```python
#Stemming

words = ['eating', 'eaten', 'eats', 'programming', 'programmed', 'programs','writing', 'written', 'writes','history','finally','finalized','congratulations']

from nltk.stem import PorterStemmer,RegexpStemmer,SnowballStemmer

ps = PorterStemmer()
for word in words:
    print(word + "-->" + ps.stem(word))

# rs = RegexpStemmer('ing$|s$|e$|able$|ly$|ed$', min=4)
# for word in words:
#     print(word + "-->" + rs.stem(word))

ss = SnowballStemmer('english')
for word in words:
    print(word + "-->" + ss.stem(word))
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')  # downloads the WordNet lemmatizer model

wnl = WordNetLemmatizer() #Noun-n Verb-v Adjective-a Adverb-r

for word in words:
    print(word + "-->" + wnl.lemmatize(word, pos='v'))
```

```python
#Stop Words
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('stopwords')  # downloads the list of stop words
nltk.download('wordnet')  # downloads the WordNet lemmatizer model
stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
wnl = WordNetLemmatizer()

description = "Climate change is one of the most pressing challenges of our time. Rising global temperatures are causing glaciers to melt, sea levels to rise, and extreme weather events to become more frequent. Governments, businesses, and individuals must work together to reduce carbon emissions and promote sustainable practices. Education and awareness play a crucial role in inspiring communities to take meaningful action for a healthier planet."
words = word_tokenize(description)
stemmed_words = [ps.stem(word) for word in words if word not in stop_words]
print(stemmed_words)
lemmatized_words = [wnl.lemmatize(word) for word in words if word not in stop_words]
print(lemmatized_words)
```

```python
#Parts Of Speech Tagging
from nltk.tag import pos_tag
tagged_words = pos_tag(words)
print(tagged_words)
```

```python
#Named Entity Recognition
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk import ne_chunk
# nltk.download('maxent_ne_chunker')  # downloads the named entity chunker model
# nltk.download('maxent_ne_chunker_tab')  # downloads the named entity chunker model for tab-separated data
# nltk.download('words')  # downloads the list of words for named entity recognition
# nltk.download('averaged_perceptron_tagger')  # downloads the POS tagger model
# nltk.download('averaged_perceptron_tagger_eng')  # downloads the English POS tagger model

sentence = "In 2021, Microsoft partnered with the University of Cambridge to launch a research initiative on quantum computing. The project was announced in London by Satya Nadella, the CEO of Microsoft, alongside Professor Emily Carter. The collaboration aimed to accelerate breakthroughs in quantum algorithms and was later supported by funding from the European Union. This initiative followed similar efforts by Google and IBM, who had already made significant progress in the field."
words = word_tokenize(sentence)
tagged_words = pos_tag(words)
entities = ne_chunk(tagged_words)
print(entities)
ne_chunk(tagged_words).draw()
```

```python
#Bag Of Words
import pandas as pd

messages = pd.read_csv('spam.csv', encoding='latin-1')
messages = messages[['v1', 'v2']] 
messages = messages.dropna()

# messages.head()

#Data Cleaning and Preprocessing
import re
# nltk.download('stopwords') 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
# nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
wnl = WordNetLemmatizer()

corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['v2'].iloc[i])
    review = review.lower().split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100)
X = cv.fit_transform(corpus).toarray()
X    
```
